```python
sel = scr.TE_Scraper(browser='firefox')
sel.load_page("https://tradingeconomics.com/united-states/non-manufacturing-pmi")
sel.click_button("a.hawk-chartOptions-datePicker-cnt-btn:nth-child(5)")
```

```python
yaxys = sel.get_y_axis()
yaxys
```

```python
sel.get_element()
series = sel.series_from_element(invert_the_series=True)
sel.series 
```

```python
sel.series.max()
```

```python
fig = sel.series.plot()

# Add horizontal lines and labels
for i in range(len(yaxys)):
    fig.add_shape(
        type='line',
        x0=sel.series.index.min(),
        y0=yaxys.index[i],
        x1=sel.series.index.max(),
        y1=yaxys.index[i],
        line=dict(color='Red', dash='dash')
    )
    fig.add_annotation(
        x=sel.series.index.max(),  # Position the label at the end of the line
        y=yaxys.index[i],
        text=str(yaxys.iloc[i]),
        showarrow=False,
        xanchor='left',
        yanchor='middle'
    )

# Show the figure
fig.show()
```
